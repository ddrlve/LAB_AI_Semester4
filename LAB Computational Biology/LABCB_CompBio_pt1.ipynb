# **Exercise from DOCX**

```python
seqA = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
seqB = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
# Find the length of both sequences
print("Length Seq A: " + str(len(seqA)))
print("Length Seq B: " + str(len(seqB)))
```

```python
# Determine the number of times the codon (CGC) appears in both sequences
print("codon(CGC) Seq A: " + str(nt_search(str(seqA), "CGC")))
print("codon(CGC) Seq B: " + str(nt_search(str(seqB), "CGC")))
```

```python
# Locate the first occurrence index position of the sub-sequence(CAGTC) in both sequences
print("sub-sequence(CAGTC) Seq A: " + str(seqA.find("CAGTC")))
print("sub-sequence(CAGTC) Seq B: " + str(seqB.find("CAGTC")))
```

```python
# Combine the first 15 Nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.
slice_seqA = seqA[:15]
slice_seqB = seqB[-10:]
print(slice_seqA)
print(slice_seqB)

seqC = slice_seqA + slice_seqB
print(seqC)
```

```python
# Reverse the sequence of sequence C.
seqC_reverse = seqC[::-1]
print(seqC_reverse)
```

```python
# Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library

freq_seqA = Counter(seqA)
print(freq_seqA)
freq_seqB = Counter(seqB)
print(freq_seqB)
freq_seqC = Counter(seqC)
print(freq_seqC)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.bar(freq_seqA.keys(), freq_seqA.values())
plt.title("Neuclotide seq A")
plt.subplot(1,3,2)
plt.bar(freq_seqB.keys(), freq_seqB.values())
plt.title("Neuclotide seq B")
plt.subplot(1,3,3)
plt.bar(freq_seqC.keys(), freq_seqC.values())
plt.title("Neuclotide seq C")
plt.show()
```
