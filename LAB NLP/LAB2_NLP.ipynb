* [v] Stemming -> mengubah sebuah kata menjadi bentuk dasarnya, dengan cara menghilangkan suffix, prefix, atau infixnya (kata imbuhan, kata gabung, dsb) ex. berjalan -> jalan
* [v] Lemmatizing -> mengubah sebuah kata menjadi bentuk dasarnya, yang membedakan adalah merubah bentuk dasarnya berdasarkan LEMMA nya (LEMMA -> kata yang udh ada di kamus, mengikuti kamus) **LEBIH AKURAT**, berdasarkan POS (part-of-speech). POS adalah pengelompokan kata yang merupakan kata sifat, kerja, dll.
* [v] Wordnet for synonyms and antonyms -> cari sinonim dan antonim dari suatu kata

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# Stemming ada 3 jenis algoritma setmming
# 1. PorterStemmer
# 2. SnowballStemmer
# 3. LancasterStemmer

from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import LancasterStemmer
# 1. PorterStemmer -> algoritmanya lambat, tapi secara hasil cukup akurat, hanya bisa stemming bahasa inggris
porter = PorterStemmer()

# 2. SnowBall -> versi yang lebih bagus dari Porter, tapi algoritmanya sama kaya porter. dia support multi language, ga cuma bahasa inggris aja, bisa stemming lebih dari 1 bahasa
snowball = SnowballStemmer('english')

# 3. LancasterStemmer -> algoritmanya cepat, tapi hasilnya buruk (akurasinya buruk)
lancaster = LancasterStemmer()
```

```python
word_list = ['dancing', 'changing', 'drawn', 'moved']

for word in word_list:
  print(f'Porter = {word} -> {porter.stem(word)}')
  print(f'Snowball = {word} -> {snowball.stem(word)}')
  print(f'Lancaster = {word} -> {lancaster.stem(word)}\n')
```

```python
# Lemmatization (Lemmatizing)
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
# simpen ke dalam sebuah variabel
lemma = WordNetLemmatizer()

word = 'better'

# Kegunaan Lemmatizing lainnya -> bisa mengubah sebuah bentuk kata menjadi bentuk kata lainnya,
# ex. better -> good (mengubah). verb->adverb. adjective -> verb, dsb.

adjective = lemma.lemmatize(word, 'a') # 'a' = adjactive

adverb = lemma.lemmatize(word, 'r') # r = adverb

verb = lemma.lemmatize(word, 'v') # 'v' = verb

noun = lemma.lemmatize(word, 'n') # 'n' = noun, ga pake 'n' juga bisa aza

print(word)
print("Adjective: " + adjective)
print("Adverb: " + adverb)
print("Verb: " + verb)
print("Noun: " + noun)
```

```python
# Mencari Synonim dan Antonim dari sebuah kata

from nltk.corpus import wordnet
nltk.download('wordnet')

from nltk.tokenize import word_tokenize
```

```python
sentence = 'Quickly kangaroos caused chaos'

word_list = word_tokenize(sentence)

# print(word_list)

for word in word_list:
  synonyms = []
  antonyms = []

  synsets = wordnet.synsets(word)  # synset merupakan kumpulan makna2 dari sebuah kata, yg berfungsi untuk mencari sinonim. 1 kata = banyak makna

  for syn in synsets:
    for lemma in syn.lemmas(): # lemma itu sebuah makna, dari kata2 yang kita ambil dari synset tadi, nnti dimasukin ke synonim tadi
      synonyms.append(lemma.name()) # lemma.name itu untuk ambil synonym nya
      for ant in lemma.antonyms():
        antonyms.append(ant.name())

  print(word)

  if synonyms:
    print(f'Synonyms: -> {synonyms[:5]}') # pake metode slicing [0:5]
  else:
    print('No Synonyms')

  if antonyms:
    print(f'Antonyms -> {antonyms[:5]}')
  else:
    print('No Antonyms')

  print()
```
