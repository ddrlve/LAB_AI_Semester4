```python
from nltk.probability import FreqDist
```

```python
sentence = "Tom told Tim to take ten tiny turtles to the tall tower, but Tim tumbled twice trying to transport the tricky turtles."
```

```python
fd = FreqDist(word_tokenize(sentence))

# print(fd) # gabisa langsung di print

for word, count in fd.most_common(10): #berapa kata yang paling banyak muncul
  print(f"{word}: {count}")
```
