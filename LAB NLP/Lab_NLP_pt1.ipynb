```python
# nltk -> natural language toolkot
# tokenizing -> memisahkan teks yang besar menjadi lebih kecil

import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentence = "I scream, you scream, we all scream for ice cream"

word_list = word_tokenize(sentence)
print(word_list)
```

```python
paragraph = "Slippery snakes slither silently. Swift sparrows spot subtle shifts. Brisk beetles bump blue berries. Clever crabs cross crowded cobblestones. Thin thrushes trill through thickets."

sentence_list = sent_tokenize(paragraph)

print(sentence_list)
```
